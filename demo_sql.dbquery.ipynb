```sql
USE parallax_demo;

WITH ranked AS (
  SELECT
    market_id,
    window_start,
    window_end,
    anomaly_score,
    volume,
    trade_count,
    imbalance,
    reason_tags,
    ROW_NUMBER() OVER (
      PARTITION BY market_id
      ORDER BY anomaly_score DESC
    ) AS market_rank,
    MAX(anomaly_score) OVER (
      PARTITION BY market_id
    ) AS market_max_score,
    AVG(anomaly_score) OVER (
      PARTITION BY market_id
    ) AS market_avg_score,
    COUNT(*) OVER (
      PARTITION BY market_id
    ) AS market_windows
  FROM gold_parallax_signals
)

SELECT
  market_id,
  window_start,
  window_end,
  anomaly_score,
  volume,
  trade_count,
  imbalance,
  reason_tags,
  market_rank,
  market_max_score,
  ROUND(market_avg_score, 2) AS market_avg_score,
  market_windows
FROM ranked
WHERE
  market_rank <= 5
  OR market_id = 'tesla_model_z_before_june_30'
ORDER BY
  market_max_score DESC,
  market_id,
  window_start;
```
